```python
print("\nNote: Currency rates are approximate and may change in real time.")
amount = float(input("Enter amount: "))

from_currency = input("Convert from (USD/INR/EUR): ")
to_currency = input("Convert to (USD/INR/EUR): ")

# USD conversions
if from_currency == "USD" and to_currency == "INR":
    result = amount * 83
    print("Amount in INR:", result)

elif from_currency == "USD" and to_currency == "EUR":
    result = amount * 0.9
    print("Amount in EUR:", result)

# INR conversions
elif from_currency == "INR" and to_currency == "USD":
    result = amount / 83
    print("Amount in USD:", result)

elif from_currency == "INR" and to_currency == "EUR":
    result = amount / 83 * 0.9
    print("Amount in EUR:", result)

# EUR conversions
elif from_currency == "EUR" and to_currency == "USD":
    result = amount / 0.9
    print("Amount in USD:", result)

elif from_currency == "EUR" and to_currency == "INR":
    result = amount / 0.9 * 83
    print("Amount in INR:", result)

# same currency
elif from_currency == to_currency:
    print("Same currency. Amount:", amount)

else:
    print("Invalid input")
```
